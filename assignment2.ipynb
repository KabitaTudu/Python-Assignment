```python
"""
Q1. Write a program to display your name 10 times using:
 (a) a for loop
 (b) a while loop
"""

# using for loop
print("Using for loop:")
for i in range(10):
    print("<name>")
    
# using while loop
print("Using while loop:")
i = 1
while i <= 10:
    print("<name>")
    i = i + 1
```

```python
"""
Q6. Write a program to display the Fibonacci series up to the n-th term.
"""

n = int(input("Enter a number:"))
first = 1
second = 1
print(first)
print(second)
for i in range(1, n + 1):
    temp = first + second
    first = second
    second = temp
    print(temp)
```

```python
"""
Q10. Write a program to verify whether the inputted number is a palindrome. (Example: For input
13231 is a palindrome but 12345 is not a palindrome)
"""

n = int(input("Enter a number:"))
temp = n
rev = 0
while temp != 0:
    ld = temp % 10
    rev = rev * 10 + ld
    temp = temp // 10
if n == rev:
    print(f"{n} is a palindrome")
else:
    print(f"{n} is not a palindrome")
```

```python
"""
Q11. Write a program that takes two numbers as input and computes their Greatest Common Divisor
(GCD).
"""

x = int(input("Enter first number:"))
y = int(input("Enter second number:"))
temp_x = x
temp_y = y
while temp_y != 0:
    temp = temp_x % temp_y
    temp_x = temp_y
    temp_y = temp
print(f"GCD of {x} and {y} = {temp_x}")
```

```python
"""
Q12. Write a program that takes two numbers as input and computes their Least Common Multiple (LCM).
"""

def gcd(a, b):
    while b != 0:
        rem = a % b
        a = b
        b = rem
    return a

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
lcm = (x * y) // gcd(x, y)
print(f"LCM OF {x} and {y} = {lcm}")
```

```python
"""
Q16. Write a Python program to evaluate the exponential series.
    i.e. e^x = 1+ X + X^2/2! + X^3/3! + ‚Ä¶‚Ä¶‚Ä¶‚Ä¶
    The program should compute the value of ùíÜ^ùíô up to a specified number of terms.
"""

x = int(input("Enter the value for x: "))
n = int(input("Enter the number of terms: "))
term = 1.0
sum_term = term
i = 1
while i <= n:
    term = term * x / i
    sum_term = sum_term + term
    i = i + 1
print(sum_term)
```

```python
"""
Q17. Write a Python program to evaluate the exponential series using its series expansion. (Taylor
series expansion)
    i.e sin(X) = X - X^3/3! + X^5/5! - ‚Ä¶‚Ä¶‚Ä¶‚Ä¶‚Ä¶‚Ä¶‚Ä¶
    The program should compute the value of sin(x) up to a specified number of terms.
"""

x = float(input("Enter the degree: "))
n = int(input("Enter the number of terms: "))

x = 3.14159 / 180.0 * x  # degree to radians
term = x
sum_sine = term
i = 1
while i < n:
    term = (- term * x ** 2) / ((2 * i) * (2 * i + 1))
    sum_sine += term
    i += 1
print(sum_sine)
```

```python
"""
Q18. Write a Python program that will evaluate the sin(x) series
    (Hints Use Taylor series expansion)
    i.e cos(X) = 1-X^2/2! + X^4/4! - X^6/6! - ‚Ä¶‚Ä¶‚Ä¶‚Ä¶‚Ä¶‚Ä¶‚Ä¶..
    The program should compute the value of cos(x) up to a specified number of terms.
"""

x = float(input("Enter the degree: "))
n = int(input("Enter the number of terms: "))

x = 3.14159 / 180.0 * x  # degree to radians
term = x
sum_cos = term
i = 1
while i < n:
    term = (- term * x ** 2) / ((2 * i) * (2 * i - 1))
    sum_cos += term
    i += 1
print(sum_cos)
```

```python
"""
Q19. Write a program to find all numbers less than a given number ùëÅ that are coprime with ùëÅ.
(Note: A pair (ùëé, ùëè)is said to be coprime if gcd (ùëé, ùëè) = 1.)
"""

def gcd(m, n):
    while n != 0:
        rem = m % n
        m = n
        n = rem
    return m

n = int(input("Enter a number: "))
for i in range(1, n):
    if gcd(i, n) == 1:
        print(i)
```

```python
"""
Q21. Write a python program to convert Decimal to Binary number system.
"""

dec = int(input("Enter a decimal number: "))
binary = ''
temp = dec
if dec == 0:
    print("0")
else:
    while temp != 0:
        rem = temp % 2
        binary += str(rem)
        temp = temp // 2
print(f'Binary equivalent of {dec} is {binary[::-1]}')
```

```python
"""
22. Write a python program to convert Binary to Decimal number system.
"""

binary = input("Enter a binary number: ")
temp = binary
dec = 0
for i in range(1, len(temp)+1):
    if temp[-i] == '1':
        dec += 2**(i-1)
print(f'Decimal equivalent of {binary} is {dec}')
```

```python
"""
Q23. Write a program that will display the pattern as follows:
    (a) Half Pyramid
    (b) Inverted Half Pyramid
    (c) Hollow Inverted Half Pyramid
    (d) Full Pyramid
    (e) Inverted Full Pyramid
    (f) Hollow Full Pyramid
"""

print("Half Pyramid")
n = int(input("Enter value for n: "))
for i in range(n):
    print('*' * (i+1))
```

```python
print("Inverted Half Pyramid")
n = int(input("Enter value for n: "))
for i in range(n, 0, -1):
    print('*' * i)
```

```python
print("Hollow Inverted Half Pyramid")
n = int(input("Enter value for n: "))
for i in range(n, 0, -1):
  for j in range(1, i+1):
    if (i == n) or (j == 1) or (j == i):
      print('*', end="")
    else:
      print(" ", end="")
  print()
```

```python
print("Full Pyramid")
n = int(input("Enter value for n: "))
for i in range(1, n+1):
  print(" " * (n-i) + "* " * (i))
```

```python
print("Inverted Full Pyramid")
n = int(input("Enter value for n: "))
for i in range(n, 0, -1):
  print(" " * (n-i) + "* " * (i))
```

```python
print("Hollow Full Pyramid")
n = int(input("Enter value for n: "))
for i in range(1, n+1):
  print(" " * (n-i), end="")
  for j in range(1, 2*n):
    if (j == 1) or (j == 2*i-1) or (i == n):
      print("*", end="")
    else:
      print(" ", end="")
  print()
```

```python
"""
Q24. Write a program that will display the number pattern as follows:
  1. Half Pyramid
  2. Inverted Half Pyramid
  3. Hollow Half Pyramid
  4. Full Pyramid
  5. Hollow Full Pyramid
  6. Hollow Inverted Half Pyramid
"""

print("Half Pyramid")
n = int(input("Enter value for n: "))
for i in range(1, n+1):
  for j in range(1, i+1):
    print(j, end=" ")
  print()
```

```python
print("Hollow Half Pyramid")
n = int(input("Enter value for n: "))
for i in range(1, n+1):
  if i == n:
    for j in range(1, n+1):
      print(j, end=" ")
  elif i == 1:
    print(1)
  else:
    print("1" + " " * (2*i-3) + str(i))
```

```python
print("Full Pyramid")
n = int(input("Enter value for n: "))
for i in range(1, n+1):
  print("  " * (n-i), end="")
  num = i
  for j in range(i):
    print(num, end=" ")
    num += 1
  num -= 2
  for j in range(i-1):
    print(num, end=" ")
    num -= 1
  print()
```
