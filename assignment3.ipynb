```python
"""
Q1. Write a Python program to sum all the items in a list. 
"""
nums = []
n = int(input("Enter the size of list: "))
print("Enter elements of the list:")
for i in range(n):
    nums.append(int(input()))
sum_nums = 0
for num in nums:
    sum_nums += num
print(f"Sum of elements of the list = {sum_nums}")
```

```python
"""
Q2. Write a Python program to get the largest number from a list.
"""
nums = []
n = int(input("Enter the size of list: "))
print("Enter elements of the list:")
for i in range(n):
    nums.append(int(input()))
great = -999
for num in nums:
    if num > great:
        great = num
print(f"Greatest number from the list = {great}")
```

```python
"""
Q3. Write a program that will create a list of N numbers and search an element using linear
search.
"""
nums = []
n = int(input("Enter the size of list: "))
print("Enter elements of the list:")
for i in range(n):
    nums.append(int(input()))
key = int(input("Enter the element to be searched in the list: "))
for i in range(len(nums)):
    if key == nums[i]:
        print(f"{key} is present at position {i+1}")
        break
else:
    print(f"{key} is not present")
```

```python
"""
Q4. Count and display the number of vowels, consonants, uppercase, lowercase characters in
string.
"""

s = input("Enter a string(only containing alphabets): ")
vowel_c, cons_c, up_c, low_c = 0, 0, 0, 0
for char in s:
    if char.lower() in ['a', 'e', 'i', 'o', 'u']:
        vowel_c += 1
    else:
        cons_c += 1
    if char.isupper():
        up_c += 1
    else:
        low_c += 1
print(f"Number of vowels = {vowel_c}")
print(f"Number of consonants = {cons_c}")
print(f"Number of uppercase characters = {up_c}")
print(f"Number of lowercase characters = {low_c}")
```

```python
"""
Q5. Input a string and determine whether it is a palindrome or not. 
"""

s = input("Enter a string: ")
rev_s = s[::-1]
if s == rev_s:
    print(f"{s} is palindrome")
else:
    print(f"{s} is not palindrome")
```

```python
"""
Q6. Program to count the number of times a character appears in a given string using a
dictionary.
"""

s = input("Enter a string: ")
char_dict = {}
for char in s:
    if char not in char_dict:
        char_dict[char] = 1
    else:
        char_dict[char] += 1
print(char_dict)
```

```python
"""
Q8. Write a program for Bubble sort.
"""
nums = []
n = int(input("Enter the size of list: "))
print("Enter elements of the list:")
for i in range(n):
    nums.append(int(input()))
for i in range(n-1):
    for j in range(n-i-1):
        if nums[j] > nums[j+1]:
            temp = nums[j]
            nums[j] = nums[j+1]
            nums[j+1] = temp
print("After bubble sort, elements of the list:")
print(nums)
```

```python
"""
Q9. Write a program for insertion sort.
"""

nums = []
n = int(input("Enter the size of list: "))
print("Enter elements of the list:")
for i in range(n):
    nums.append(int(input()))
for i in range(1, n):
    temp = nums[i]
    j = i -1
    while j >= 0 and temp < nums[j]:
        nums[j+1] = nums[j]
        j -= 1
    nums[j+1] = temp
print("After insertion sort, elements of the list:")
print(nums)
```

```python
"""
Q10. Write a program that will merge sorted two list. [after merge operation the final list
should be sorted]
"""

nums1, nums2 = [], []
n1 = int(input("Enter the length of list 1: "))
print("Enter elements of the list 1 in sorted order:")
for i in range(n1):
    nums1.append(int(input()))

n2 = int(input("Enter the length of list 2: "))
print("Enter elements of the list 2 in sorted order:")
for i in range(n2):
    nums2.append(int(input()))

#  Sorted merged list
nums_merged = []

# Store all the elements in the merged list
for num in nums1:
    nums_merged.append(num)
for num in nums2:
    nums_merged.append(num)

# Sort the unsorted merged list
for i in range(1, n1+n2):
    key = nums_merged[i]
    j = i - 1
    while j >= 0 and nums_merged[j] > key:
        nums_merged[j+1] = nums_merged[j]
        j -= 1
    nums_merged[j+1] = key

print(nums_merged)
```

```python
"""
Q11. Write a program that will read a string and display the words of the string in reverse
order.
Example: myString="Silicon Institute of Technology"
Output: Technology of Institute Silicon
"""
string = input("Enter a string: ")
words = string.split()
rev_str = ''
for i in range(len(words)):
    rev_str += words[-i-1] + ' '
print(rev_str)
```

```python
"""
Q12. Write a program to print all elements in a list those have only single occurrence. 
"""
elem_list = []
n = int(input("Enter the length of the list: "))
print("Enter the elements of the list: ")
for i in range(n):
    elem_list.append(input())
print("Elements having one occurrence in the list: ")
for elem in elem_list:
    if elem_list.count(elem) == 1:
        print(elem, end=" ")
```

```python
"""
Q13. Write a Python program to find the list of words that are longer than n from a given list
of words. 
"""
n1 = int(input("Enter the length of the list: "))
words_list = []
print("Enter the elements of the list: ")
for i in range(n1):
    words_list.append(input())
n = int(input("Enter the value of n: "))
words_n = []
for word in words_list:
    if len(word) > n:
        words_n.append(word)
print(f"List of words that are longer than {n}: ")
print(words_n)
```

```python
"""
Q14. Write a program to input roll numbers and their names of students of your class and
store them in the dictionary as the key-value pair. Perform the following operations on the
dictionary:
a. Display the Roll numbers and name for all students.
b. Add a new key-value pair in this dictionary and display the modified
dictionary
c. Delete a particular student's record from the dictionary
d. Modify the name of an existing students
"""

stu_dict = {}
n = int(input("Enter the number of students: "))
for i in range(n):
    roll = int(input(f"Enter roll number of student {i+1}: "))
    name = input(f"Enter name of student {i+1}: ")
    stu_dict[roll] = name
# display roll and name of all students
print(stu_dict)
# addition of another student in the dictionary
stu_dict[4] = 'Megha'
# delete a particular student
roll_del = int(input("Enter roll number of the student to be deleted: ")) 
stu_dict.pop(roll_del)
# modification of a student data
stu_dict[4] = 'Supriya'

# final dictionary
print(stu_dict)
```

```python
"""
Q15. Write a double list comprehension over the lists ['A','B','C'] and [1,2,3] whose value is the list of
all possible two-element lists [letter, number].
That is, the value is [['A', 1], ['A', 2], ['A', 3], ['B', 1], ['B', 2],['B', 3], ['C', 1], ['C', 2], ['C', 3]]
"""
# l1 = ['A', 'B', 'C']
# l2 = [1, 2, 3]
l1 = input("Enter letters(separated by a space): ").split()
l2 = list(map(int, input("Enter numbers(separated by a space): ").split()))
res = [[x, y] for x in l1 for y in l2]
print(res)
```

```python
"""
Q21. Given a dictionary that may contain nested dictionaries as values, write Python programs to
flatten the dictionary so that all keys appear in a single-level dictionary.
Input: nested_dict = { "name": "Rahul", "age": 25, "address": { "city": "Bhubaneswar", "state":
"Odisha", "zip": { "code": 751001, "area": "Cuttack Road" } }, "hobbies": ["reading", "painting"] }
Output: flattened_dict = {"name": "Rahul", "age": 25, "address.city": "Bhubaneswar",
"address.state": "Odisha", "address.zip.code": 751001, "address.zip.area": "Cuttack Road",
 "hobbies": ["reading", "painting"]}
"""
nested_dict = {"name": "Rahul", "age": 25, "address": {"city": "Bhubaneswar", "state":
"Odisha", "zip": {"code": 751001, "area": "Cuttack Road"}}, "hobbies": ["reading", "painting"]}
flat_dict = {}
for info in nested_dict:
    if info == "address":
        flat_dict["address.city"] = nested_dict[info]["city"]
        flat_dict["address.state"] = nested_dict[info]["state"]
        flat_dict["address.zip.code"] = nested_dict[info]["zip"]["code"]
        flat_dict["address.zip.area"] = nested_dict[info]["zip"]["area"]
    else:
        flat_dict[info] = nested_dict[info]
        
print(flat_dict)
```
