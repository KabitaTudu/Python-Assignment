```python
"""
Q1. Create a tuple with different data and create an empty tuple using built-in function.
"""
tup1 = (1, 'one', 1.00, {1: ['apple', 'orange'], 2: ['cabbage', 'beans']})
print(tup1)
# creation of tuple using built-in function
tup2 = tuple()
print(tup2)
```

```python
"""
Q2. Print the 3rd element from beginning and 4th element from the end of the tuple.
"""
tup3 = tuple(map(int, input("Enter elements of a tuple: ").split()))
print(f"3rd element from beginning of tuple: {tup3[2]}")
print(f"4th element from end of tuple: {tup3[-4]}")
```

```python
"""
Q3. Count the number of items present in a tuple and then add two new items into it(one item using concatenation 
operator and another item using built-in function).
"""
tup4 = (11, 22, 33, 44, 55)
print(f"Number of elements present in tuple: {len(tup4)}")
# using concatenation operator
tup4 = tup4 + (66, 77)
print(tup4)
# using built-in function
tup4 = list(tup4)
tup4.extend([88, 99])
print(tuple(tup4))
```

```python
"""
Q4. Find the number of occurrences for any repeated item present in the tuple.
"""
tup5 = tuple(map(int, input("Enter elements of a tuple: ").split()))
repeat_dict = {}
for elem in tup5:
    if tup5.count(elem) > 1 and elem not in repeat_dict:
        repeat_dict[elem] = tup5.count(elem)
print(repeat_dict)
```

```python
"""
Q5. Display the elements present in a tuple which are at the even index position.
"""
tup6 = tuple(map(int, input("Enter elements of a tuple: ").split()))
for i in range(0, len(tup6), 2):
    print(tup6[i], end=" ")
```

```python
"""
Q6. Remove the elements of a tuple which are present in the odd index position.
"""
tup7 = tuple(map(int, input("Enter elements of a tuple: ").split()))
lst_tup7 = list(tup7)
odd_tup7 = ()
for i in range(0, len(lst_tup7), 2):
    odd_tup7 += (lst_tup7[i],)
print(odd_tup7)
```

```python
"""
Q7. Remove the odd number elements from a tuple.
"""
tup8 = tuple(map(int, input("Enter elements of a tuple: ").split()))
lst_tup8 = list(tup8)
removed = lst_tup8.copy()
for elem in lst_tup8:
    if elem % 2 != 0:
        removed.remove(elem)
tup8 = tuple(removed)
print(tup8)
```

```python
"""
Q8. Reverse the elements of a tuple by using and without using a built-in function.
"""
# using built-in function
tup9 = (3, 2, 6, 8, 3, 5)
tup9 = list(tup9)
tup9.reverse()
print(tuple(tup9))

# without using built-in function
tup10 = (5, 2, 7, 1, 1, 9, 0)
rev_tup10 = []
for i in range(len(tup10)-1, -1, -1):
    rev_tup10.append(tup10[i])
print(tuple(rev_tup10))
```

```python
"""
9. Sort the elements present in a tuple in descending order.
"""
tup11 = tuple(map(int, input("Enter elements of a tuple: ").split()))
desc_tup11 = tuple(sorted(tup11, reverse=True))
print(desc_tup11)
```

```python
"""
10. Take a tuple of mixed data types and find the square of each element.
"""
tup12 = (3, 6, 1.5, 2.0, 'one', 4)
sq_tup12 = tuple(x**2 for x in tup12 if isinstance(x, (int, float)))
print(sq_tup12)
```

```python
"""
11. Given few numbers in a tuple, return ‘True’ if the first and last numbers are same.
"""
tup13 = tuple(map(int, input("Enter elements of a tuple: ").split()))
if tup13[0] == tup13[-1]:
    print(True)
else:
    print(False)
```

```python
"""
12. Print only those elements present in a tuple which are divisible by 5.
"""
tup14 = tuple(map(int, input("Enter elements of a tuple: ").split()))
for elem in tup14:
    if elem % 5 == 0:
        print(elem, end=" ")
```

```python
"""
13. Given a tuple of numbers, create another tuple such that it contains only the even
numbers of the first tuple.
"""
tup15 = tuple(map(int, input("Enter elements of a tuple: ").split()))
even_tup15 = tuple()
for elem in tup15:
    if elem % 2 == 0:
        even_tup15 += (elem,)
print(even_tup15)
```

```python
"""
14. Display the largest number of a tuple without using the built-in function max().The
user need to input the values in tuple from the keyboard.
"""
tup16 = tuple(map(int, input("Enter elements of a tuple: ").split()))
max_elem = tup16[0]
for elem in tup16:
    if elem > max_elem:
        max_elem = elem
print("Largest number:", max_elem)
```

```python
"""
15. Replace the last value of the tuple with 100.
Sample list: [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
Output: [(1, 2, 100), (4, 5, 100), (7, 8, 100)]
"""
sample_lst = list(map(tuple, input("Enter some tuples: ").split(',')))
lst_100 = []
for elem in sample_lst:
    lst_100.append(elem[:-1] + (100,))
print(lst_100)
```

```python
"""
16. Sort a tuple by its float element.
Sample data: [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]
Output: [('item3', '24.5'), ('item2', '15.10'), ('item1', '12.20')]
"""
sample_data = []
n = int(input("Enter the size of list: "))
print("Enter the tuple of string and float:")
for i in range(n):
    item_str = input("Enter the string: ")
    item_float = float(input("Enter the float: "))
    sample_data += [(item_str, item_float)]
sorted_tup = tuple(sorted(sample_data, key=lambda x:x[1], reverse=True))
print(sorted_tup)
```

```python
"""
17. Count the number of elements in a tuple until an element is a list.
"""
c = 0
tup17 = (3, 12, [3, 5], 87, 0)
for elem in tup17:
    if isinstance(elem, list):
        break
    c += 1
print(c)
```

```python
"""
18. Swap the two tuples: tuple1 = (11, 22) tuple2 = (99, 88)
Output: tuple1 = (99, 88) tuple2 = (11, 22)
"""
tup1 = tuple(map(int, input("Enter elements of tuple1: ").split()))
tup2 = tuple(map(int, input("Enter elements of tuple2: ").split()))
tup1, tup2 = tup2, tup1
print(f"Tuple1 = {tup1} Tuple2 = {tup2}")
```
