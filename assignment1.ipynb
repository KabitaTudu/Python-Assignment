```python
# Q3. Write a program to find the largest of three numbers using nested condi􀆟onal operators.

num1, num2, num3 = int(input("Enter first number:")), int(input("Enter second number:")), int(input("Enter third number:"))
largest = 0
if num1 > num2:
  if num1 > num3:
    largest = num1
  else:
    largest = num3
else:
  if num2 > num3:
    largest = num2
  else:
    largest = num3

print("Largest:", largest)
```

```python
"""
Q4. Write a Python program to display the grade of a student based on the marks obtained according
to the following criteria:
 Marks greater than or equal to 90 → O Grade
 Marks greater than or equal to 80 and less than 90 → E Grade
 Marks greater than or equal to 70 and less than 80 → A Grade
 Marks less than 70 → Fail
"""

marks = int(input("Enter marks:"))
grade = ''
if marks >= 90:
  grade = 'O'
elif marks >= 80:
  grade = 'E'
elif marks >= 70:
  grade = 'A'
else:
  grade = 'Fail'

print("Grade:", grade)
```

```python
"""
Q5. Write a program to calculate the electricity bill by taking previous and current meter readings
and applying the following rules:
 Units consumed = Current Reading - Previous Reading
 First 100 units → ₹1/unit
 Next 100 units → ₹2/unit
 Remaining units → ₹3/unit
"""

prev_reading = float(input("Enter previous meter reading:"))
curr_reading = float(input("Enter current meter reading:"))
units_consumed = curr_reading - prev_reading
if units_consumed >= 200:
  bill = 3*(units_consumed - 200) + 200 + 100
elif units_consumed >= 100:
  bill = 2*(units_consumed - 100) + 100
else:
  bill = 0.00

print("Electricity bill:", bill)
```

```python
"""
Q6. Write a program to calculate the net salary of an employee based on basic salary using the
following rules:
 HRA = 20% of basic salary
 DA = 10% of basic salary
 Net Salary = Basic + HRA + DA
"""

basic_salary = float(input("Enter basic salary:"))
net_salary = basic_salary + .2*basic_salary + .1*basic_salary
print("Net salary:", net_salary)
```

```python
"""
Q7. Write a Python program to calculate the income tax payable by an individual based on their
annual income according to the following tax slabs:
 Up to ₹2,50,000 → No tax
 ₹2,50,001 to ₹5,00,000 → 5% tax
 Above ₹5,00,000 → 10% tax
"""

income = float(input("Enter income:"))
if income >= 500000:
  tax = 250000 + .1*(income - 500000)
elif income > 250000:
  tax = .05*(income - 250000)
else:
  tax = 0.00

print("Income tax payable:", tax)
```

```python
"""
Q8. Write a program, that will read the principal amount and other necessary parameter, compute
the simple interest and compound interest.
"""

principal = float(input("Enter principal amount:"))
rate = float(input("Enter rate of interest:"))
time = float(input("Enter time:"))
si = (principal * time * rate) / 100
ci = (principal*(1 + rate / 100)**time) - principal
print("Simple interest:", si)
print("Compound interest:", ci)
```

```python
"""
Q9. Write a program which read the radius of the circle as input and print its area and perimeter.
"""

radius = float(input("Enter radius of a circle:"))
pi = 3.14
area = pi*radius**2
perimeter = 2*pi*radius
print(f"Area: {area:.2f}")
print(f"Perimeter: {perimeter:.2f}")
```

```python
"""
Q16. Write a program to check whether a given number is a power of two using bitwise operators.
"""

num = int(input("Enter a number:"))
if ( num > 0) and (num & (num - 1) == 0):
  print(num, "is a power of 2")
else:
  print(num, "is not a power of 2")
```

```python
"""
Q17. Write a program that will toggle (flip) the kᵗʰ bit of a number.
"""

num = int(input("Enter a number:"))
k = int(input("Enter the value for k:"))
print(num ^ (1 << k))
```

```python
"""
Q18. Given a number you have to check whether there is pair of adjacent set bit or not.
  Example-1:
    Input : N = 67
    Output : Yes
    There is a pair of adjacent set bit
    The binary representation is 100011
  Example-2:
    Input : N = 5
    Output : No
"""

num = int(input("Enter a number:"))
if (num & (num >> 1) == 0):
  print("No")
else:
  print("Yes")
```

```python
"""
https://www.hackerrank.com/challenges/py-if-else/problem?isFullScreen=true

Given an integer, n, perform the following conditional actions:

If n is odd, print Weird
If n is even and in the inclusive range of 2 to 5, print Not Weird
If n is even and in the inclusive range of 6 to 20, print Weird
If n is even and greater than 20, print Not Weird

Input Format
A single line containing a positive integer, n.

Constraints
1 <= n <= 100

Output Format
Print Weird if the number is weird. Otherwise, print Not Weird.
"""

n = int(input())
if n % 2 != 0:
  print("Weird")
else:
  if n > 20 or n in range(2, 6):
    print("Not Weird")
  else:
    print("Weird")
```
